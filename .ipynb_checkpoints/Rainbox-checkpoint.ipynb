```python
import json
```

```python
with open("curiosity_data.json") as file:
    info =  json.load(file)
```

```python
info
```

```python
l = []
for ele in info:
#     print(ele['data'])
    for data in ele['data']:
        for sub in data['subcategories']:
            d = {}
            d['category'] = data['name']
            d['sub-category'] = sub['name']
            d['img_url'] = sub['image_url']
            l.append(d)
```

```python
l
```

```python
print(len(l))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(l)
```

```python
df
```

```python
df.to_csv("movies.csv")
```

```python
df.to_json("movies.json")
```
